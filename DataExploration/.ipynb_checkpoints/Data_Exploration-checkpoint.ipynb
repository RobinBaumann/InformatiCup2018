# Data Exploration


Gain some insight into the dataset using Pandas, Numpy and Scikit Learn

```python
%matplotlib inline
import matplotlib.pyplot as plt
import os
import pandas as pd
import numpy as np

DATASET_PATH = "../Eingabedaten/"
GAS_PRICES = "../Eingabedaten/Benzinpreise/"

gas_stations_with_prices_id_list = [int(os.path.splitext(os.path.basename(os.path.join(GAS_PRICES, f)))[0]) for f in os.listdir(os.path.join(GAS_PRICES)) if os.path.isfile(os.path.join(GAS_PRICES, f))]

gas_stations = pd.DataFrame()

def load_gas_station_data(data_location=DATASET_PATH):
    gas_stations = os.path.join(data_location, "Tankstellen.csv")
    return pd.read_csv(gas_stations, sep=";", names=["id", "name", "brand", "street", "house_number", "zip_code", "city", "latitude", "longitude"])


def load_historic_prices(gas_station_id, data_location=DATASET_PATH):
    path = data_location + "Benzinpreise/{}.csv".format(gas_station_id)
    historic_prices = os.path.join(path)
    return pd.read_csv(historic_prices, sep=";", header=None)
```

```python
gas_stations = load_gas_station_data()
clean_gas_stations = gas_stations[gas_stations['id'].isin(gas_stations_with_prices_id_list)]
clean_gas_stations.reset_index()
clean_gas_stations.head()
```

```python
def calculate_median_prices(gas_stations_id_list):
    median_prices = [load_historic_prices(x)[1].median() for x in gas_stations_id_list]
    return median_prices
```

```python
medians = calculate_median_prices(sorted(gas_stations_with_prices_id_list))
```

```python
means_df = pd.DataFrame(medians, index=sorted(gas_stations_with_prices_id_list))
data = pd.concat([clean_gas_stations, means_df], axis=1)
data.head()
```

```python
ax = data.plot(kind="scatter", 
                       x="longitude", 
                       y="latitude", 
                       alpha=0.2,
                       figsize=(10,10),
                       c=0,
                       cmap=plt.get_cmap("OrRd"),
                       colorbar=True,
                       vmin=min(medians),
                       vmax=max(medians)
                      )

ax.set_xlabel("Längengrad")
ax.set_ylabel("Breitengrad")
```

```python
from collections import defaultdict

#brands from https://benzinpreis.de/statistik.phtml
important_brands = ["Agip", "Aral", "Avia", "Esso", "bft", "Freie", "HEM", "JET", "Oili", "OMV", "Raiffe", "SBTank", "Shell", "star", "Supermarkt"]

def prices_of_station(station_id):
    historic_prices = load_historic_prices(station_id)
    return historic_prices[1].mean()
    

def group_brands(resulting_brands, gas_stations):
    stations = defaultdict(list)
    for index, row in gas_stations.iterrows(): 
        for r in resulting_brands:
            if r.lower() not in str(row['brand']).lower():
                continue
            else:
                stations[r].append(prices_of_station(int(row['id'])))
    return stations

brands = group_brands(important_brands, clean_gas_stations)
brands_mean = {}
for key in brands:
    brands_mean[key] = np.mean(brands[key])
```

```python
for k in brands_mean:
    print('{}: {}'.format(k, brands_mean[k]))
```

```python
ax = plt.figure(figsize=(15,5))

ax = plt.bar(range(len(brands_mean)), brands_mean.values(), align='center')
plt.xticks(range(len(brands_mean)), brands_mean.keys())

plt.ylim(1200,1400)

plt.show()
```

```python
import time
import datetime
def getPricesInTime(id,start,end):
    data = load_historic_prices(id)
    data[0] = data[0].apply(lambda row: row.split("+")[0])
    data[0] = data[0].apply(lambda row: time.mktime(datetime.datetime.strptime(row, "%Y-%m-%d %H:%M:%S").timetuple()))
    sequence = data.loc[((data[0] > start) & (data[0] < end))]
    return sequence

step = (time.time()-1.402214e+09)/20
x_val = []

means = [];
for x in range(1,int(20),1):
    temp_data = getPricesInTime(12304,(1.402214e+09)+step*(x-1),(1.402214e+09)+step*x)
    means.append(temp_data[1].mean())
    x_val.append(datetime.datetime.fromtimestamp(temp_data[0].min()))
    
x_val
```

```python
import matplotlib.pyplot as plt
import matplotlib.dates as md
plt.plot(x_val, means, 'ro')
ax=plt.gca()
xfmt = md.DateFormatter('%Y-%m-%d %H:%M')
ax.xaxis.set_major_formatter(xfmt)
plt.subplots_adjust(bottom=0.2)
plt.xticks( rotation=25 )
plt.show()
```
