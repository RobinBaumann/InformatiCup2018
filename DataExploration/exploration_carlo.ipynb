# Aufbau Tankstellen Daten

FÃ¼r das Clustering der Tankstellen ist wichtig welche Tankstelle in welchem  Zeitraum existiert.

```python
import pandas as pd
con = create_connection()

df = pd.read_sql_query("""
select id, min_ts, max_ts
from stations
where min_ts is not null""", con=con)
df

```

```python
min_tss = df['min_ts'].copy()
min_tss = pd.Series(min_tss.values.astype('<M8[D]'))
min_tss.groupby(min_tss).count()
```

```python
max_tss = df['max_ts'].copy()
max_tss = pd.Series(max_tss.values.astype('<M8[D]'))
max_tss.groupby(max_tss).count()
```

Zwischen 2015-01-01 und 2017-09-18 sollten genug Tankstellen durchgehend Daten haben, um weitere Analysen durchzufÃ¼hren.

```python
active_stations = pd.read_sql_query("""
select id, min_ts, max_ts
from stations
where min_ts <= '2015-01-01'
and max_ts >= '2017-09-18'""", con)
active_stations
```

13615 Tankstellen sind zu viel, um verschieden Clustering Verfahren miteinander zu vergleichen und noch dieses Jahr ein Ergebnis zu erwarten. Deshalb werden die Ids der in Frage kommenden Tankstellen gesampled.

```python
import numpy as np
ids = np.random.choice(active_stations['id'], 100)
ids
```

```python
param = ','.join([str(x) for x in ids])
query = """select * from prices_sampled where station_id in (%s) 
and time_stamp between '2015-01-01' and '2017-09-18'""" % param
prices = pd.read_sql_query(query, con)
prices
```

```python
prices_pivoted = prices.pivot_table(index='time_stamp', columns='station_id', values='price')
prices_pivoted
```
