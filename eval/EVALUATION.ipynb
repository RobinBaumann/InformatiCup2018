```python
import matplotlib.pyplot as plt
import numpy as np

trees = [0, 3, 7, 12, 17, 22, 27, 33, 40, 47, 50]
rmse = [9.05291, 7.55989, 6.33699, 5.53512, 5.12321, 4.86165, 4.66613, 4.50698, 4.37921, 4.29096, 4.25628]


mpl_fig = plt.figure(figsize=(5, 2), dpi=300)
ax = mpl_fig.add_subplot(111)

plt.plot(trees, rmse, 'ro-')
plt.axis([-1, 51, 4, 10])

ax.set_xlabel("Entscheidungsbäume")
ax.set_ylabel("RMSE [cent/Liter]")

plt.show()
```
